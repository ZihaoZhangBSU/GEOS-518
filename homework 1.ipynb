# Part 1:

## Initial parameters and packages

```python
# import neccessary package for processing and rename them for convenience
import numpy as np   #for data processing  
import matplotlib.pyplot as plt   # for plotting

Nt = 200    # The number of time steps of walkers
Nx = 1000   # The number of walkers 

mu_x = 1.5  # The average step size in our random walk
s_x = 2.1    # The standard deviation of the step size in our random wal
min_x = 1.25 # The minimum step size at each step in the random walk
```

## Create the random walk 2-d array

```python
# initial an array X to store the how long each walker walk every time step
X = np.zeros((Nt,Nx))

for i in np.arange(Nt):

    # The foor loop below loops through each walker. It is 'nested' within the 
    # time for loop
    for j in np.arange(Nx):
        x_i = 0.0 # set an initial value for xi to meet the cycle requirements
        while x_i <= min_x:  # judge if xi is equal or less than min_x
        # Calculate the random step size for the j^th walker at the i^th time step
            x_i = s_x*np.random.randn() + mu_x
        # Again check to see if we're at the first time step
        if(i==0):
            # If at the first time step, the initial position was 0
            X[i,j] = 0.0 + x_i 
        else:
            # If at a subsequent time step, the initial position was the position at the previous time step
            X[i,j] = X[i-1,j] + x_i 
```

## Calculate statistics 

```python
mean_X = np.mean(X,axis=1) # Calculate the average distance of all walkers through time
std_X = np.std(X,axis=1) # Calculate the standard deviation of all walkers through time
```

## Plot figure 5.1

```python
plt.figure(figsize=(16,10)) # set the figure size
# set the variables and legends for these plots below
plt.plot(np.arange(Nt),X,'b-')  
plt.plot(np.arange(Nt),mean_X,'k-')
plt.plot(np.arange(Nt),mean_X + std_X, 'k--')
plt.plot(np.arange(Nt),mean_X - std_X, 'k--')
plt.xlabel('Number of Steps') 
plt.ylabel('Distance Traveled [m]') 
```

# Part 2:

```python
Nt = 300  # The number of time steps of walkers
Nx = 3    # The number of walkers 

mu_x = 0.5  # The average step size of x direction 
s_x = 1.5  # The standard deviation of the step size of x direction 
mu_y = 5.0  # The average step size of y direction 
s_y = 2.5  # The standard deviation of the step size of y direction  
```

```python
# initial an array X to store the how long each walker walk every time step
X = np.zeros((Nt,2, Nx))
# use the first dimension for x location, and second dimension for y location
for i in np.arange(Nt):

    # The foor loop below loops through each walker. It is 'nested' within the 
    # time for loop
    for j in np.arange(Nx):
        # Calculate the random step size for the j^th walker at the i^th time step
        x_i = s_x*np.random.randn() + mu_x
        y_i = s_y*np.random.randn() + mu_y
        # Again check to see if we're at the first time step
        if(i==0):
            # If at the first time step, the initial position was 0
            X[i,0, j] = 0.0 + x_i   
            X[i,1, j] = 0.0 + y_i
        else:
            # If at a subsequent time step, the initial position was the position at the previous time step
            X[i,0, j] = X[i-1,0, j] + x_i 
            X[i,1, j] = X[i-1,1, j] + y_i  
```

```python
plt.figure(figsize=(10, 6))

plt.plot(X[:,0, 0], X[:,1, 0], linestyle='-', color='b', label='Person 1')
plt.plot(X[:,0, 1], X[:,1, 1], linestyle='-', color='g', label='Person 2')
plt.plot(X[:,0, 2], X[:,1, 2], linestyle='-', color='r', label='Person 3')
```
